```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 6, 100)
y_1 = np.exp(x)
y_2 = 10 * x + 20

plt.figure(figsize=(5, 6), dpi=120)
plt.rcParams.update({'font.size': 12})
plt.plot(x, y_1, lw=3, label='Using Event Instances')
plt.plot(x, y_2, lw=3, label='Centralise Event Manager')
plt.xticks([])
plt.yticks([])
plt.xlabel('Number of Objects')
plt.ylabel('Complexity')
plt.legend(loc=2)
# plt.grid()
plt.show()
```
